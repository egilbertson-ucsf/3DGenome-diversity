# Find outlier regions

```python
num_shuffles=10000
s = pd.read_csv('%s/differentiation_shuffles/basic_10000_shuffles_diffs_set2.csv' % RESULTS_PATH, index_col=[0,1], )#names=range(1,num_shuffles+1))

s = s.dropna(axis=1, how='all')
save = s.copy()
s.columns = names=range(1,num_shuffles+1)
```

```python
s
```

```python
s.loc['chr20']
```

```python
s20 = s.loc['chr20']
s20.index = s20.index.map(int)
```

```python
level_to_change = 1
s.index = s.index.set_levels(s.index.levels[level_to_change].astype(str), level=level_to_change)
s['minimum'] = s.min(axis=1)
s['maximum'] = s.max(axis=1)
shuff_range = s[['minimum','maximum']]
```

```python
pop_diffs = pd.DataFrame()
for pop in pops.keys():
    pop_diffs[pop] = pops[pop]['new_fst']
```

```python
pop_diffs = pd.concat([shuff_range, pop_diffs], axis=1)
```

```python
pop_diffs
```

```python
for pop in pops.keys():
    pop_diffs['%s_out'%pop] = (pop_diffs[pop].between(pop_diffs.minimum, pop_diffs.maximum)==False)
```

```python
print(pop_diffs[pop_diffs['AFR_out']].shape)
pop_diffs[pop_diffs['AFR_out']]
```

```python
pop_diffs[pop_diffs['EAS_out']]
```

```python
pop_diffs[pop_diffs['SAS_out']]
```

```python
pops_out=pop_diffs[['AFR_out', 'AMR_out','EUR_out','EAS_out','SAS_out']]
```

```python
pops_out = pops_out[(pops_out ==True).any(axis=1)]
```

```python
pops_out.loc['chr20']['SAS_out']
```

```python
indicators = pops_out.loc['chr20'].reset_index().drop(columns=['index'])
```

```python
indicators
```

```python
listy = pops_out.loc['chr20'].index
listx=[]
for i in listy:
    listx.append(str(i))
```

```python
data = pd.DataFrame({"value": listx, **indicators})
```

```python
from upsetplot import plot, from_indicators
```

```python
ind_dict = {}
for pop in indicators.columns:
    ind_dict[pop] = list(indicators[pop].values)
```

```python
indicators = ind_dict
```

```python
e = from_indicators(ind_dict, data = data)
```

```python
fig= plt.figure(figsize=(20,10))
plot(e, fig=fig, element_size=None, sort_by='cardinality', show_counts=True)
fig.savefig('popsout_focalVall_chr20.svg',format='svg')
```
